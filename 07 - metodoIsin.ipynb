## Importando o pandas

```
import pandas as pd
```

## Lendo o Data Frame

```
df1 = pd.DataFrame({'A': [3,2,5], 'B':[1,2,3], 'C':[4,5,6]})
df2 = pd.DataFrame({'A': [4,5,'a'], 'B':[2,3,4], 'C':['b',6,5]})
```

##Visualizando o Data Frame 1

## O método isin pode filtrar , retorna a comparação de Data Frames por booleano

##Visualizando com o isin

```
df2[df2['B'].isin(['2', '3'])]
```

```
df1.isin([1, 2, 3])
```

```
valores = ['a', 'b', 1, 3]
df2.isin(valores)
```
