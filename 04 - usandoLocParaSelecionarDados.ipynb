## Importando o pandas e o leitor de url

```
import pandas as pd
import io
import requests
```

## Lendo a url do arquivo csv

```
url = "https://raw.githubusercontent.com/dressalsl/tutorialPandas/master/dataSets/brazil-microcephaly-2016-01-23-table-1.csv"
df = pd.read_csv(url, sep=',')
df.head()
```

## loc = Este método é baseado nas labels (legenda) da colunas
### Usando loc com índice



```
df.loc[[4]]
```

## Usando loc com array

```
df.loc[[0,4,18]]
```

## Usando loc por fatiamento

```
df.loc[5:9]
```

##Usando o loc para visualizar a última linha

```
df.loc[-1]
```

####Causa erro, pois procura a linha que tem a legenda/tag/rótulo de índice (nº da linha) de -1 e não existe. Vamos manipular com python.

```
n_linhas = df.shape[0]
n_linhas
```

```
ultima_linha = n_linhas -1
df.loc[ultima_linha]
```

## Para usar o loc chamando pelo nome da linha, devemos primeiro setar um index

```
df2 = df.set_index("state")
```

## Visualizando a alteração

### Após setar, nossa label são os estados

```
df2.head()
```

##Vizualizando apenas Pernambuco

```
df2.loc["Pernambuco"]
```

## Podemos selecionar linhas e colunas usando um array com linhas e outro com as colunas

```
df2.loc[["Pernambuco","Ceará","Bahia"],["cases_confirmed","cases_reported_total"]]
```

## Filtrando casos confirmados acima de 10 

```
df.loc[df['cases_confirmed'] > "10"]
```
