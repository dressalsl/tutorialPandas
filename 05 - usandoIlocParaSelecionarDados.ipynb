## Importando o pandas e o leitor de url

```
import pandas as pd
import io
import requests
```

## Lendo a url do arquivo csv

```
url = "https://raw.githubusercontent.com/dressalsl/tutorialPandas/master/dataSets/brazil-microcephaly-2016-01-23-table-1.csv"
df = pd.read_csv(url, sep=',')
df.head()
```

## iloc =  selecionar dados de linhas e colunas pelo seu de índice.

---



## Selecionando a primeira linha do DataFrame

```
df.iloc[0]
```

## Selecionando a última linha do DataFrame

```
df.iloc[-1]
```

## Selecionando todos os dados da segunda coluna

```
df.iloc[:,1]
```

## Selecionando do 1º ao 5º da última coluna

```
df.iloc[0:5,-1]
```

## Selecionando as três primeiras linhas do DataFrame

```
df.iloc[0:3]
```

## Selecionando todos os dados da 2º e da última coluna

```
df.iloc[:,[1,5]]
```

## Selecionando a 1º, a 3º e 6º linha da 1º até a 3º coluna

```
df.iloc[[0,2,5], 1:4]
```
