```
# Com um dicionário iremos criar o nosso DataFrame para plotar os gráficos
df = pd.DataFrame({'faltas' : [3,4,2,1,4,1],
                   'prova' : [10,7,5,8,6,3],
                   'seminario': [8.5,7.5,9.0,7.5,8.0,7.0],
                   }, index = ["felipe", "hugo", "paulo", "luis", "vitor","pedro"])
df
```

```
# Plota o gráfico valores x index (nesse caso são os nomes dos alunos), por default o tipo é line
df.plot()
# Plota o gráfico do tipo linha e com a cor definida
df.plot(kind='line', color="hotpink")
```

```
# Plotando um tipo de coluna e definindo sua cor
df["prova"].plot.bar(color="tomato") 
```

```
# Plota separadamente cada gráfico, ou seja, subplots
df.plot.line(subplots=True)
```

```
# Plota o gráfico no formato de barra
df.plot(kind='bar')
```

```
# Plota com valores que forem definidos para os eixos x e y, tipo: barra
df.plot(kind='bar',x='faltas',y='prova')
# tipo: dispersão 
df.plot(kind='scatter',x='faltas',y='prova')
```

```
# Plotando em formato de pizza e definindo o seu tamanho
df.plot.pie(y='prova', figsize=(8,8))
```

```
# Plotando subplots em formato de pizza
df.plot.pie(subplots=True, figsize=(30, 30))
```
