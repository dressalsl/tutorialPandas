## Importando o pandas e o leitor de url

```
import pandas as pd
import io
import requests
```

## Lendo a url do arquivo csv

```
url = "https://raw.githubusercontent.com/dressalsl/tutorialPandas/master/dataSets/bank.csv"
df = pd.read_csv(url, sep=';')
df.head()
```

## O metódo .groupby() serve para agrupar os dados para fazer operações e manipulações com eles

## Com o .groupby().mean() tiramos a média , nesse caso da coluna "marital"

```
df.groupby('marital').mean()
```

## Tirando a média de várias colunas com SubSets. Observe "duration" é do tipo int64 para poder tirar a média

```
  df[["marital", "contact", "duration"]].groupby(["marital", "contact"]).mean()
```

## Com o .groupby().mean().sort_values() sabemos os valores da média de determinada coluna em ordem crescente

```
df.groupby("marital").mean()["age"].sort_values()
```

## Com o .groupby().first() apresenta o primeiro dos valores dentro de cada grupo 

```
df.groupby(['marital', 'age']).first().head(10)
```

## Média das idades das pessoas com idade entre 18 e 50 anos com a escolaridade primaria e secundaria

```

colunas = ["education","age"]
df[((df.age >= 18) & (df.age <= 50)) & (df.education.isin(["primary", "secondary"]))][colunas].groupby(["education"]).mean()
```
