```
# Importando o pandas
import pandas as pd
```

```
url = "https://raw.githubusercontent.com/dressalsl/tutorialPandas/master/usandoSelecionadores/bank.csv"
df = pd.read_csv(url, sep=';')
df.head()
```

```
# O metódo groupby serve para agrupar os dados com informação das médias agrupadas
```

```
df.groupby('marital').mean()
```

```
# Para saber os valores da média de determinada coluna em ordem crescente
df.groupby("marital").mean()["day"].sort_values()
```

```
df.groupby(['marital', 'duration']).first().head(10)
```

```
df[["marital", "contact", "duration"]].groupby(["marital", "contact"]).mean()
```

```
# Média das idades das pessoas com idade entre 18 e 50 anos coma escolaridade primaria e secundaria
colunas = ["age", "education"]
df[((df.age >= 18) & (df.age <= 50)) & (df.education.isin(["primary", "secondary"]))][colunas].groupby(["education"]).mean()
```
